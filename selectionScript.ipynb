```python
keepCols = df.columns[4:]
allCols = ['metric','metric ticker','metric calcrt','model combo']
skipCols = allCols

allCols += [i+"_"+str(j) for j in range(1,4) for i in keepCols]
#print allCols

dataDict = {str(k): []  for k in allCols}
# print dataDict
for row in df.groupby(['metric','model combo']):
    #print row
    #print len(row[1])
    for i in range(len(row[1])+1,4):
        for col in row[1]:
            if col not in skipCols:
                dataDict[col+"_"+str(i)].append(None)
        
    for col in row[1]:
        #print len(row[1])
        #print row[1][col].iloc[cIndex]
        if col in skipCols:
            dataDict[col].append(row[1][col].iloc[0])
        else:
            for cIndex in range(len(row[1])):
                dataDict[col+"_"+str(cIndex+1)].append(row[1][col].iloc[cIndex])
        

# for k,v in dataDict.iteritems():
#     print k, len(v)
finalFrame = pd.DataFrame(dataDict)[allCols]
```
